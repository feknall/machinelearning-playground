# Neural Network for AND

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential

train_x = [[0, 0], [0, 1], [1, 0], [1, 1]]
train_y = [0, 0, 0, 1]

model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(2, )))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(train_x, train_y, epochs=300)

loss, acc = model.evaluate(train_x, train_y)
print(loss, acc)
```

# SVM for XOR Gate

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

train_x = [[0, 0], [0, 1], [1, 0], [1, 1]]
train_y = [0, 1, 1, 0]

my_svm = svm.SVC()
my_svm.fit(train_x, train_y)
predict_y = my_svm.predict(train_x)

score = my_svm.score(train_x, train_y)
print(score)

confu_matrix = confusion_matrix(train_y, predict_y)
print(confu_matrix)

class_report = classification_report(train_y, predict_y)
print(class_report)

accuracy_score = accuracy_score(train_y, predict_y)
print(accuracy_score)
```

# Logistic Regression for OR

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

train_x = [[0, 0], [0, 1], [1, 0], [1, 1]]
train_y = [0, 1, 1, 1]

logistic_regression = LogisticRegression(penalty='none')
logistic_regression.fit(train_x, train_y)
predict_y = logistic_regression.predict(train_x)

score = logistic_regression.score(train_x, train_y)
print(score)

confu_matrix = confusion_matrix(train_y, predict_y)
print(confu_matrix)

class_report = classification_report(train_y, predict_y)
print(class_report)

accuracy_score = accuracy_score(train_y, predict_y)
print(accuracy_score)
```
